```python
def fit(X, Y):
    return np.linalg.solve(X.T.dot(X), X.T.dot(Y))
```

```python
def make_poly(X, deg):
    n = len(X)
    data = [np.ones(n)]
    for d in X range(deg):
        data.append(X ** (d+1))
        return np.vstack(data).T
```
